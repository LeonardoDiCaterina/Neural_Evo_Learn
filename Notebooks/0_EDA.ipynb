## 5. Outliers

```python
data_no_target = data_for_ploting.copy()
data_no_target.drop(columns=['CRUDE PROTEIN'], inplace=True)
```

```python
idx_of_outliers = []
for col in data_no_target.columns:
    above_99 = data_no_target[col]>= data_no_target[col].quantile(0.99)
    idx_of_outliers.append(data_no_target[above_99].index.tolist())

idx_of_outliers = [item for sublist in idx_of_outliers for item in sublist]
idx_of_outliers_set = set(idx_of_outliers)
idx_of_outliers_set
```

```python
idx_outlier_count = pd.Series(idx_of_outliers).value_counts()
print(idx_outlier_count)
print(f"Number of rows with outliers: {idx_outlier_count.sum(axis=0)}")
```

```python
# Show the row with the most outliers
data_no_target.loc[[75]]
```

> Outliers Removing by IQR range

```python
#calculate the interquartile range (IQR) for each column
outliers_df = pd.DataFrame()
for col in data_no_target.columns:
    q1 = data_no_target[col].quantile(0.25)
    q3 = data_no_target[col].quantile(0.75)
    iqr = (q3 - q1)
    # take the rows that are outside the bounds
    above_upper_bound = data_no_target[col] > (q3 + 1.5*iqr)
    below_lower_bound = data_no_target[col] < (q1 - 1.5*iqr)
    outliers = data_no_target[(below_lower_bound) | (above_upper_bound)]
    # store the outliers in a dataframe
    outliers_df = pd.concat([outliers_df, outliers], axis=0)
    # drop the outliers
    df_no_outlier_iqr = data_no_target.drop(outliers.index)
```

```python
# check if there are duplicated rows in the outliers dataframe
print(f"Duplicated outliers: {outliers_df.duplicated().sum(axis=0)}")
print(f"Number of outliers identified by IQR method: {outliers_df.shape[0]}")
print(f"Percentage of outliers identified by IQR method: {(outliers_df.shape[0]/data.shape[0])*100:.2f}%")
```
